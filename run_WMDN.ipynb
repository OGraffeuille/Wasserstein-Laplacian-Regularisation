```python
import pandas as pd
from datetime import datetime

# Our scripts
from src.WMDN import *
from src.utils import *
```

```python
# User-selected parameters
data_name = "UCI_air_quality"
X, y = get_data(data_name)

l, u, v = 300, 2700, 1000
t = np.minimum(X.shape[0]-4000, 10000)
n_seeds = 3
seed_start = 0
num_hidden = [32, 32, 32]
lr = 1e-4
ssl_weight = 1

# Generate parameters
input_dims = X.shape[1]
seeds = list(range(seed_start, seed_start+n_seeds))
is_ssl = (u > 0)

# Store results
MDN_results_df = pd.DataFrame()

# Begin experiments
for seed in seeds:
    
    # Train model
    model = WMDN()
    model.fit(X, y, l, u, v, t, \
              input_dims=input_dims, num_hidden=num_hidden, \
              seed=seed, lr=lr, ssl_weight=ssl_weight)
    
    # Plot results
    model.plot_results()
    
    # Evaluate metrics
    metrics_v = model.get_metrics(model.X_t_ten, model.y_t_ten, verbose=True)
    metrics_t = model.get_metrics(model.X_t_ten, model.y_t_ten, verbose=True)
    
    # Record results
    row = {
        "data_name": data_name,
        "l": l,
        "u": u, 
        "v": v,
        "t": t,
        "ssl_weight": ssl_weight,
        "num_hidden": " ".join(str(l) for l in num_hidden),
        "lr": lr,
        "seed": seed,
        "epochs": model.epoch
    }
    for metric in metrics_t:
        row[metric+"_v"] = metrics_v[metric]
        row[metric+"_t"] = metrics_t[metric]
    MDN_results_df = MDN_results_df.append(row, ignore_index=True)


# Save results dataframe
filename_suffix = "{} {} IsSSL;{}".format(datetime.now().strftime('%Y%m%d%H%M%S'), data_name, is_ssl)
print(" ########## Saving: "+ filename_suffix)
MDN_results_df.to_csv("results/" + filename_suffix + ".csv")
```
